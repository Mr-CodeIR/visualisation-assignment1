```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
#load data from sklearn
iris = datasets.load_iris()
X = iris.data
y = iris.target#as we have categorical data we need to use targets to  to work with categorical data
#managing figure size
fig = plt.figure(1, figsize=(8, 4))
#showing 3-d plot and managing how it looks to user
ax = Axes3D(fig,elev=38, azim=138)
#to center data points we use PCA
pca = decomposition.PCA(n_components=3)
pca.fit(X,y)
X = pca.transform(X)#Apply dimensionality reduction to X.
#placing labels for datapoints and matching them with datapoints
for name, label in [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]:
    ax.text3D(X[y == label, 0].mean(),
              X[y == label, 1].mean() + 1.5,
              X[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=.5, facecolor='w'))
# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y)


plt.show()
```
